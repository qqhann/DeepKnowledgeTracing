```python
with open('./data/assistments/builder_train.csv', 'r') as f:
    lines = f.readlines()
```

```python
lines
```

```python
i = 0
all_questions = set()
while i < len(lines):
    # nStepsStr
    steps = int(lines[i].strip())
    if steps < 2:
        i += 3
        continue
    i += 1
    # questionIdStr
    questions = list(map(int, lines[i].strip().rstrip(',').split(',')))
    all_questions |= set(questions)
    
    i += 1
    # correctStr
    i += 1
```
